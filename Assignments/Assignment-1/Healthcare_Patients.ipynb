```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(font_scale=1.05)
plt.style.use("ggplot")
```

```python
df = pd.read_csv("patients.csv")

print("Raw Data Sample:")
df.head()
```

```python
# Drop missing / invalid
df = df.dropna(subset=["id", "name", "age", "gender", "medical_expense"])
df = df.drop_duplicates(subset="id")

# Fix gender
df["gender"] = df["gender"].astype(str).str.upper().str[0]

# Ensure numeric
df["age"] = pd.to_numeric(df["age"], errors="coerce")
df["medical_expense"] = pd.to_numeric(df["medical_expense"], errors="coerce")
df = df.dropna(subset=["age", "medical_expense"])
```

```python
df["expense_category"] = pd.cut(
    df["medical_expense"],
    bins=[0, 1000, 5000, 10000, 20000, 50000],
    labels=["Very Low", "Low", "Medium", "High", "Very High"],
    include_lowest=True
)
df.head()
```

```python
print("Basic Statistics:\n", df["medical_expense"].describe())
print("\nExpense Category Counts:\n", df["expense_category"].value_counts())
```

```python
# 1. Age distribution
plt.figure(figsize=(8,5))
sns.histplot(df["age"], bins=20, kde=True)
plt.title("Distribution of Patient Ages")
plt.show()

# 2. Expense distribution
plt.figure(figsize=(8,5))
sns.histplot(df["medical_expense"], bins=30, kde=True, color="green")
plt.title("Distribution of Medical Expenses")
plt.show()

# 3. Expense categories count
plt.figure(figsize=(7,5))
sns.countplot(x="expense_category", data=df, order=["Very Low","Low","Medium","High","Very High"])
plt.title("Patients per Expense Category")
plt.show()

# 4. Total expense by gender
plt.figure(figsize=(6,5))
sns.barplot(x="gender", y="medical_expense", data=df, estimator=sum, ci=None, palette="pastel")
plt.title("Total Medical Expenses by Gender")
plt.show()
```

```python
df.to_csv("patients_cleaned.csv", index=False)
print(" Cleaned data saved: patients_cleaned.csv")
```
